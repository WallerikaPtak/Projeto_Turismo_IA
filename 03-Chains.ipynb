```python
# CÉLULA PARA O ARQUIVO: 03-Chains.ipynb
# Responsabilidade: Demonstrar as cadeias simples (roteiro, logística, tradução).

import os
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(temperature=0, model_name="gemma2-9b-it")

# Cadeia de Roteiro
prompt_roteiro = PromptTemplate.from_template("Crie um roteiro de viagem para: {input}")
chain_roteiro = prompt_roteiro | llm | StrOutputParser()
print("--- Teste Roteiro ---")
print(chain_roteiro.invoke({"input": "3 dias em Paris para um casal"}))

# Cadeia de Logística
prompt_logistica = PromptTemplate.from_template("Responda à pergunta de logística de viagem: {input}")
chain_logistica = prompt_logistica | llm | StrOutputParser()
print("\n--- Teste Logística ---")
print(chain_logistica.invoke({"input": "Como ir do aeroporto de Guarulhos para a Avenida Paulista?"}))
```
